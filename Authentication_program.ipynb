```python
import random

#registeration
Registered_email = input("Enter your email: ")
secret_question = input("Enter a secret question: ")
secret_answer = input("Enter your secret answer: ")

#generate random password
rand_char = ['a','b','c','d','e','f','g','h','i','j','k','l','m','n','B','A','C','D','E','F','G','H','I','J','1','2','3','4','5','6','7','8','9','0','/','?','!','#','%','*','_']
generated_password = []
for i in range(20):
    c = random.randint(0, len(rand_char)-1)
    a  = rand_char[c]
    generated_password.append(a)
registered_password = "".join(generated_password)
print(f'Your password is: {registered_password}')

#change password
change_password = input("Do you want to change your password. Enter y or n: ")
if change_password == 'y':
    current_password = input("Enter your current password: ")
    if current_password == registered_password:
        new_password = input("Enter a new password: ")
        registered_password = new_password
    else:
        print('You are not Authorized!!')
elif change_password == 'n':
    print("thank you!")
else:
    print("Enter a valid input")

#login
email = input("Enter your registered email: ")
password = input("Enter your registered password: ")

if email == Registered_email:
    if password == registered_password:
        print('You are in!!')
    else:
        for password_tries in range(5):
            if password != registered_password:
                if password_tries < 4:
                    password = input('try again!: ')
                else:
                    print('too many password tries, reset yout password')  
                    #reset password
                    print(secret_question)
                    reset_answer = input("Enter your secret answer: ")
                    if secret_answer == reset_answer:
                        reset_password = input("Enter a new password: ")
                        registered_password = reset_password
                        email = input("Enter your email: ")
                        password = input("Enter your password: ")
                        if email == Registered_email:
                            if password == registered_password:
                                print('You are in!!')
                            else:
                                print('Incorrect email or password')
                        else:
                            print('Incorrect email or password')
                    else:
                        print("verification failed!")
            else:
                print('incorrect password!!')
                break   
else:
    print("Please enter a valid email or password")
```
